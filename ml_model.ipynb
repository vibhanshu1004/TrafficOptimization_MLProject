```python
traffic_data['date_time'] = pd.to_datetime(traffic_data['date_time'], format='%Y-%m-%d %H:%M:%S')

traffic_data['year'] = traffic_data['date_time'].dt.year
traffic_data['month'] = traffic_data['date_time'].dt.month
traffic_data['day'] = traffic_data['date_time'].dt.day
traffic_data['hour'] = traffic_data['date_time'].dt.hour
traffic_data['day_of_week'] = traffic_data['date_time'].dt.dayofweek  # Monday=0, Sunday=6

traffic_data.drop('date_time', axis=1, inplace=True)
```

```python
traffic_data = traffic_data[traffic_data['temperature'] != 0]
```

```python
#Categorizing days based on weekday
def day_of_week(day) -> str:

    if day in [0, 1, 2, 3, 4]:
        return "weekday"
    else:
        return "weekend"

traffic_data["weekday_categorical"] = traffic_data["day_of_week"].map(day_of_week)
```

```python
traffic_data.head(100)
```

```python
traffic_data.drop(columns=["day_of_week"], inplace=True)
```

```python
# Categorizing hours like morning, night, etc.
def hour(hour_value) -> str:  
    if hour_value in [4, 5, 6, 7]:
        return "early_morning"
    elif hour_value in [8, 9, 10, 11]:
        return "morning"
    elif hour_value in [12, 13, 14, 15]:
        return "afternoon"
    elif hour_value in [16, 17, 18, 19]:
        return "evening"
    elif hour_value in [20, 21, 22, 23]:
        return "night"
    else:
        return "late_night"  

traffic_data["hour_categorical"] = traffic_data["hour"].map(hour)
```

```python
traffic_data.drop(columns=["hour"], inplace=True)
```

```python
traffic_data.head(100)
```

```python
# Function to classify rain and snow presence
def snowrain_division(data) -> str:  
    if data > 0:
        return "yes"
    else:
        return "no"

traffic_data["rain_p_h"] = traffic_data["rain_p_h"].map(snowrain_division)
traffic_data["snow_p_h"] = traffic_data["snow_p_h"].map(snowrain_division)
```

```python
traffic_data.head(100)
```

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  
    Q3 = df[column].quantile(0.75)  
    IQR = Q3 - Q1  

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
    return outliers

columns_to_check = [
    "air_pollution_index", "wind_speed", "humidity", "wind_direction", 
    "visibility_in_miles", "dew_point", "temperature", "clouds_all"
]

for col in columns_to_check:
    outliers = detect_outliers_iqr(traffic_data, col)  
    print(f"Outliers in '{col}': {list(outliers)}")  

    traffic_data = traffic_data.drop(index=outliers)

traffic_data = traffic_data.reset_index(drop=True)

print("Dataset after removing outliers:")
print(traffic_data)
```

```python
traffic_data.loc[traffic_data['is_holiday'] != 'No Holiday', 'is_holiday'] = 1
traffic_data.loc[traffic_data['is_holiday'] == 'No Holiday', 'is_holiday'] = 0
traffic_data['is_holiday'] = traffic_data['is_holiday'].astype(int)
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
traffic_data['weather_type'] = le.fit_transform(traffic_data['weather_type'])
traffic_data['weather_description'] = le.fit_transform(traffic_data['weather_description'])
traffic_data['weekday_categorical'] = le.fit_transform(traffic_data['weekday_categorical'])
traffic_data['hour_categorical'] = le.fit_transform(traffic_data['hour_categorical'])

traffic_data
```

```python
traffic_data['rain_p_h'] = le.fit_transform(traffic_data['rain_p_h'])
traffic_data['snow_p_h'] = le.fit_transform(traffic_data['snow_p_h'])
```

```python
X = traffic_data.drop(columns=['traffic_volume'])
y = traffic_data['traffic_volume']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.neural_network import MLPRegressor

mlp_model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000,learning_rate_init=0.001, random_state=42)
mlp_model.fit(X_train, y_train)

y_pred_mlp = mlp_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("Mean Squared Error (MSE): " + str(mean_squared_error(y_test, y_pred_mlp)))
print("Mean Absolute Error (MAE): " + str(mean_absolute_error(y_test, y_pred_mlp)))
r2 = r2_score(y_test, y_pred_mlp)
print(f"RÂ² Score: {r2}")
```

```python
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

from keras.models import Sequential
from keras.layers import LSTM, Dense

lstm_model = Sequential()
lstm_model.add(LSTM(50, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

lstm_model.fit(X_train_lstm, y_train, epochs=50, batch_size=32)
```

```python
y_pred_lstm = lstm_model.predict(X_test_lstm)

mae_lstm = mean_absolute_error(y_test, y_pred_lstm)
r2_lstm = r2_score(y_test, y_pred_lstm)

print(f'LSTM Model - MAE: {mae_lstm}, R^2 Score: {r2_lstm}')
```
