```python
traffic_data['date_time'] = pd.to_datetime(traffic_data['date_time'], format='%Y-%m-%d %H:%M:%S')

traffic_data['year'] = traffic_data['date_time'].dt.year
traffic_data['month'] = traffic_data['date_time'].dt.month
traffic_data['day'] = traffic_data['date_time'].dt.day
traffic_data['hour'] = traffic_data['date_time'].dt.hour
traffic_data['day_of_week'] = traffic_data['date_time'].dt.dayofweek  # Monday=0, Sunday=6

traffic_data.drop('date_time', axis=1, inplace=True)
```

```python
#Categorizing days based on weekday
def day_of_week(day) -> str:

    if day in [0, 1, 2, 3, 4]:
        return "weekday"
    else:
        return "weekend"

traffic_data["weekday_categorical"] = traffic_data["day_of_week"].map(day_of_week)
```

```python
traffic_data
```

```python
# Categorizing hours like morning, night, etc.
def hour(hour_value) -> str:  
    if hour_value in [4, 5, 6, 7]:
        return "early_morning"
    elif hour_value in [8, 9, 10, 11]:
        return "morning"
    elif hour_value in [12, 13, 14, 15]:
        return "afternoon"
    elif hour_value in [16, 17, 18, 19]:
        return "evening"
    elif hour_value in [20, 21, 22, 23]:
        return "night"
    else:
        return "late_night"  

traffic_data["hour_categorical"] = traffic_data["hour"].map(hour)
```

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  
    Q3 = df[column].quantile(0.75)  
    IQR = Q3 - Q1  

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
    return outliers

columns_to_check = [
    "air_pollution_index", "wind_speed", "humidity", "wind_direction", 
    "visibility_in_miles", "dew_point", "temperature", "clouds_all"
]

for col in columns_to_check:
    outliers = detect_outliers_iqr(traffic_data, col)  
    print(f"Outliers in '{col}': {list(outliers)}")  

    traffic_data = traffic_data.drop(index=outliers)

traffic_data = traffic_data.reset_index(drop=True)

print("Dataset after removing outliers:")
print(traffic_data)
```

```python
traffic_data.loc[traffic_data['is_holiday'] != 'No Holiday', 'is_holiday'] = 1
traffic_data.loc[traffic_data['is_holiday'] == 'No Holiday', 'is_holiday'] = 0
traffic_data['is_holiday'] = traffic_data['is_holiday'].astype(int)
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
traffic_data['weather_type'] = le.fit_transform(traffic_data['weather_type'])
traffic_data['weather_description'] = le.fit_transform(traffic_data['weather_description'])
traffic_data['weekday_categorical'] = le.fit_transform(traffic_data['weekday_categorical'])
traffic_data['hour_categorical'] = le.fit_transform(traffic_data['hour_categorical'])

traffic_data
```

```python
X = traffic_data.drop(columns=['traffic_volume'])
y = traffic_data['traffic_volume']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.1, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
mape =mean_absolute_percentage_error(y_test, y_pred)
print("Mean Absolute Percentage Error:", mape)
```

```python
from sklearn.neural_network import MLPRegressor

mlp_model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=100, random_state=42)
mlp_model.fit(X_train, y_train)

y_pred_mlp = mlp_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print("Mean Absolute Error (MAE): " + str(mean_absolute_error(y_test, y_pred_mlp)))
r2 = r2_score(y_test, y_pred_mlp)
print(f"RÂ² Score: {r2}")
```

```python
import numpy as np

def create_sequences(data, target, window_size):
    sequences, labels = [], []
    for i in range(len(data) - window_size):
        sequences.append(data[i:i+window_size])
        labels.append(target[i + window_size])
    return np.array(sequences), np.array(labels)

window_size = 5

X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train_scaled, window_size)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test_scaled, window_size)
```

```python
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization
from keras.optimizers import Adam

lstm_model = Sequential([
    Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
    LSTM(128, return_sequences=True, activation='tanh', dropout=0.2),
    BatchNormalization(),
    LSTM(64, return_sequences=False, activation='tanh', dropout=0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

# Compile the model
lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
```

```python
import keras_tuner as kt

def build_model(hp):
    model = Sequential([
        Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        LSTM(hp.Int('units_1', min_value=32, max_value=128, step=32), 
             return_sequences=True, dropout=hp.Float('dropout_1', 0.1, 0.5, step=0.1)),
        BatchNormalization(),
        LSTM(hp.Int('units_2', min_value=32, max_value=128, step=32),
             return_sequences=False, dropout=hp.Float('dropout_2', 0.1, 0.5, step=0.1)),
        Dense(hp.Int('dense_units', min_value=16, max_value=64, step=16), activation='relu'),
        Dropout(hp.Float('dropout_dense', 0.1, 0.5, step=0.1)),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=hp.Choice('lr', [0.001, 0.0005, 0.0001])), 
                  loss='mse', metrics=['mae'])
    return model

tuner = kt.RandomSearch(
    build_model,
    objective='val_mae',
    max_trials=10,
    executions_per_trial=1,
    directory='my_dir',
    project_name='lstm_tuning'
)

tuner.search(X_train_seq, y_train_seq, epochs=50, validation_split=0.2, batch_size=32)
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = lstm_model.fit(
    X_train_seq, y_train_seq,
    validation_split=0.2,
    epochs=100,
    batch_size=64,
    callbacks=[early_stopping]
)
```

```python
from sklearn.model_selection import train_test_split

X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train_seq, y_train_seq, test_size=0.2, random_state=42
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

y_pred = lstm_model.predict(X_test_seq)

y_pred_original = scaler.inverse_transform(y_pred)
y_test_original = scaler.inverse_transform(y_test_seq)

mae = mean_absolute_error(y_test_original, y_pred_original)
mape = mean_absolute_percentage_error(y_test_original, y_pred_original)

print(f"MAE: {mae}, MAPE: {mape}")
```
